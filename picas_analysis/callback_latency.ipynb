```python
from caret_analyze.record import Latency
from caret_analyze import Application, Architecture, Lttng

arch = Architecture('yaml', './architecture_path.yaml')
lttng = Lttng('/home/atsushi22/.ros/tracing/autoware_launch_trace_20230308-062832')
app = Application(arch, lttng)
```

```python
import numpy as np

for cb in app.callbacks:
    records = Latency(cb.to_records()).to_records()
    latency_list = records.get_column_series(records.columns[1])
    if not latency_list: continue
    latency_np = np.array(latency_list)
    print(f'|{cb.callback_name}| max: {round(latency_np.max()*10**(-6), 4)}, 93: {round(np.percentile(latency_np, 93)*10**(-6), 4)}')
```
